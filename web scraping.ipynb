### Grabbing Keith Galli Image and Storing it on a file in my local computer

```python
url="https://keithgalli.github.io/web-scraping/webpage.html"
req = requests.get(url)
```

```python
#getting the content of the webpage
get_content  = bs(req.content)
print(get_content)
```

```python
# getting all image links
img = get_content.select("img")
img_list = [get_img["src"] for get_img in img ]
print(img_list)
```

```python
# this function cleans the link and makes is appendable
def clean_img_list(picture_list): # picture_list is the list of the links to the picture
    counter = 0
    pattern = re.compile("^[\.|\/|\^]*") # matches the pattern for me
    pic_list = picture_list
    for iter_pic in pic_list:
        match = pattern.search(iter_pic)  # searches the characters occuring at the beginning of the link and gets the span
        match_find = pattern.findall(iter_pic)  # finds the character that is matched
        if match_find[0] != "": # if the first item that the regex pattern matches is not empty slice it out
            pic_list[counter] = iter_pic[match.span()[1]:] # slices the remaining portion of the link and reassigns it
        counter = counter + 1
    return pic_list
clean_img_list(img_list)
```

```python
#attaching the links to the url
from urllib.parse import urljoin
def final_url(func, url_list): # func is the returned element of the clean_img_list 
    fun_list = func  # assigns the function to a new variable
    url_list = url    #assigns the url to the webpage
    correct_url = []
    for get_func_url in fun_list:
        correct_url.append(urljoin(url, get_func_url)) # joins the two url together to get the final url
    return correct_url
final_url(clean_img_list(img_list), url)
```
