```python
import re
from bs4 import BeautifulSoup as bs
import requests
```

```python
# get_url link
url = "https://keithgalli.github.io/web-scraping/webpage.html"
```

```python
# get the requests
req = requests.get(url)
soup = bs(req.content)
print(soup.prettify())
```

```python
# extract the links with images
get_img = soup.select("img")
img_list = [get_image["src"] for get_image in get_img]
img_list
```

```python
# cleaning the link
def url_cleaner(image_list):
  image_list = image_list
  counter = 0
  for get_new_img in image_list:
    pattern = re.compile("^[\.|\^|\*|\#|\/]*")
    find_match = pattern.findall(get_new_img)  # locates all part of the link that matches the pattern
    match = pattern.search(get_new_img)  # gives me more info about the items
    if find_match != "":
      image_list[counter] = get_new_img[match.span()[1]:] # slices the other part of the link
    counter = counter + 1
  return image_list
print(url_cleaner(img_list))
      
```

```python
# concatinating the link
from urllib.parse import urljoin
def final_url(url_cleaner, img_list, url_link): # calls the cleaner func and passes its argument and also the url
  url_cleaner_func = url_cleaner(img_list)
  final_url_list = []
  for get_url in url_cleaner_func:
    final_url_list.append(urljoin(url_link, get_url))
  return final_url_list
final_link_list = final_url(url_cleaner, img_list, url)
```

```python
# reading the image
img_list = [requests.get(get_img).content for get_img in final_link_list] # getting the contents of the image link
for read_link in img_list:
  with open("image{}.jpg".format(img_list.index(read_link)), "wb") as fimage: # reading the content as image file
    fimage.write(read_link)
```
